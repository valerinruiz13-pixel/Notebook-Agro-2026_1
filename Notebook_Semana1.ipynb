```python
# ENTRAD: Entra ancho y alto de la madera en metros
ancho = input("Ingresa el ancho de la ventana en metros:")
ancho= float(ancho)
alto = input("Ingresa el alto de la ventana en metros:")
alto= float(alto)

# PROCESO: Calcula longitud de la madera (lm) en pies
# y el Area de vidrio en metros cuadrados
lm= 2 * (ancho + alto) * 3.25
av= 2 * (ancho * alto)

# Salida: Sale la longitud de la madera en pies
# el area del vidrio en metros cuadrados
print("Longitud de la madera es ", lm, " pies de longitud")
print("El area del vidrio es ", av, " metros cuadrados")

```
