#**Capacidad de Silos de Granos**


Este ejercicio permite aplicar variables tipo float y operaciones matemáticas para el cálculo de volúmenes y almacenamiento en sistemas agroindustriales.

**Planteamiento:**


Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico.
El usuario debe ingresar el radio(r) y la altura(h) del silo en metros, para determinar su volumen y posteriormente la cantidad de toneladas de maíz almacenado, considerando la densidad del material.

**Requerimientos:**

1. Solicitar el radio del silo en metros (Tipo float) (r).
2. Solicitar la altura del silo en metros (Tipo float) (h).
3. Considerar la densidad del maíz como 0.75 ton/m³.
4. Calcular el volumen del silo:

$$V = \pi r^2 h$$

5. Calcular la cantidad de toneladas almacenadas:

$$ton = V \cdot 0.75$$

**Prueba del programa:**

Si se ingresan 3 m de radio y 10 m de altura, el programa debe calcular que el silo puede almacenar aproximadamente 212.06 toneladas de maíz.

```python
# ENTRADA: Solicitar el radio y la altura del silo (Tipo float).
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

# PROCESO: Calcular el volumen del silo y las toneladas de maíz
import math
densidad = 0.75  # ton/m^3

volumen = math.pi * (radio ** 2) * altura
toneladas = volumen * densidad

# SALIDA: Informar la capacidad del silo
print("La capacidad del silo es de", round(toneladas, 2), "toneladas de maíz")
```
