#**Rendimiento de Pulpa en Frutas**

Este ejercicio es ideal para aplicar variables tipo float y operaciones aritméticas en el cálculo de rendimiento en procesos agroindustriales.

**Planteamiento:**

Una planta procesadora necesita calcular el porcentaje de aprovechamiento de la fruta. El usuario debe ingresar el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas), para determinar la cantidad de pulpa obtenida y su rendimiento.

**Requerimientos:**

1. Solicitar el peso total de la fruta en kg (Tipo float). (pt)
2. Solicitar el peso de los residuos en kg (Tipo float). (r)
3. Calcular la cantidad de pulpa:

$$Pulpa = pt - r$$

4. Calcular el porcentaje de rendimiento:

$$\%Rendimiento = \left(\frac{Pulpa}{pt}\right) \cdot 100$$

**Prueba del programa:**

Si se ingresan 500 kg de fruta y 150 kg de residuos, el programa debe calcular:

Pulpa = 350 kg  
Rendimiento = 70%

```python
# ENTRADA: Solicitar el peso total y los residuos (Tipo float).
peso_total = float(input("Ingrese el peso total de la fruta (kg): "))
residuos = float(input("Ingrese el peso de los residuos (kg): "))

# PROCESO: Calcular la pulpa y el rendimiento
pulpa = peso_total - residuos
rendimiento = (pulpa / peso_total) * 100

# SALIDA: Informar los resultados
print("La cantidad de pulpa es:", pulpa, "kg")
print("El rendimiento es:", round(rendimiento, 2), "%")
```
