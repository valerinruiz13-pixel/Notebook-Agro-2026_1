## ¿Qué es Python?

En términos sencillos, Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser "legible": leer código en Python es casi como leer instrucciones en un manual.

## 3 Ejemplos de Código (Sencillos y Útiles)

Aquí tienes tres scripts básicos que podrías aplicar en un entorno agroindustrial.

### 1. Cálculo de Rendimiento de Extracción

Este código ayuda a calcular qué tan eficiente fue un proceso de extracción (por ejemplo, de aceite de palma o jugo de naranja).

```
# Datos de entrada
materia_prima = 500  # kg
producto_obtenido = 120  # kg

# Cálculo del rendimiento
rendimiento = (producto_obtenido / materia_prima) * 100

print(f"El rendimiento de la extracción es del {rendimiento}%")

if rendimiento < 20:
    print("Alerta: El rendimiento es bajo. Revisar presión de los equipos.")
```

### 2. Clasificación Automática de Calidad (Grado)

Imagina un sensor que mide el Grado Brix (azúcar) de una fruta. Este código clasifica el lote automáticamente según estándares comerciales.

```
grado_brix = 12.5  # Dato obtenido de un refractómetro

if grado_brix >= 14:
    categoria = "Premium (Exportación)"
elif grado_brix >= 10:
    categoria = "Estándar (Mercado local)"
else:
    categoria = "Industrial (Para pulpa o mermelada)"

print(f"Resultado del análisis: Categoria {categoria}")
```

### 3. Estimación de Necesidad de Riego

Un pequeño modelo lógico basado en la humedad del suelo detectada por un sensor para optimizar el uso del agua.

```
humedad_suelo = 18  # Porcentaje de humedad detectado por sensor
punto_marchitez = 20 # Umbral crítico para el cultivo

print(f"Humedad actual: {humedad_suelo}%")

if humedad_suelo < punto_marchitez:
    print("ESTADO: Suelo seco. Activando electroválvulas de riego...")
else:
    print("ESTADO: Humedad óptima. Sistema de riego en espera.")
```
