#**Dosificación de Fertilizante Líquido**

Este ejercicio es ideal para aplicar variables tipo float y operaciones aritméticas en el cálculo de dosificación en sistemas de riego agrícola.

**Planteamiento:**

Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea. El usuario debe ingresar las hectáreas totales y la dosis requerida (litros por hectárea) para calcular la cantidad total de fertilizante y el volumen de agua necesario para la dilución.

**Requerimientos:**

1. Solicitar la cantidad de hectáreas (Tipo float). (ha)
2. Solicitar la dosis de fertilizante en L/ha (Tipo float). (d)
3. Calcular la cantidad total de fertilizante:

$$producto = ha \cdot d$$

4. Considerar la relación de dilución 1:500 (1 L de producto por 500 L de agua).
5. Calcular la cantidad de agua necesaria:

$$agua = producto \cdot 500$$

**Prueba del programa:**

Si se ingresan 5 hectáreas y una dosis de 2 L/ha, el programa debe calcular:

Producto = 10 L  
Agua = 5000 L

```python
# ENTRADA: Solicitar las hectáreas y la dosis de fertilizante (Tipo float).
hectareas = float(input("Ingrese la cantidad de hectáreas: "))
dosis = float(input("Ingrese la dosis (L/ha): "))

# PROCESO: Calcular la cantidad de producto y agua necesaria
producto = hectareas * dosis
agua = producto * 500

# SALIDA: Informar la dosificación total
print("Se necesitan", producto, "litros de fertilizante")
print("Se necesitan", agua, "litros de agua")
```
